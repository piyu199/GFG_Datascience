### Slicing in Dataframe using loc and iloc

```python
data={
    "one" : [1,2,3,4],
    "two" : [10,20,30,40],
    "three" : [100,200,300,400],
    "four": [1000,2000,3000,4000] 
}
df=pd.DataFrame(data,index=["a","b","c","d"])
df
```

```python
# loc() function is label based data selecting method which means we have to pass the name of row/column which
#  we want to select. This method include the last element of the range passed 
print(df.loc["b"])
print("-"*30)
print(df.loc["b":"d","two":"four"])
print("-"*30)
print(df.loc[["b","d"],["two","four"]]) 
```

```python
df
```

```python
# iloc() function is index based data selecting method which means we have to pass the index of row which
# we want to select.This method does not include the last element of the range passed 
df.iloc[0:3,0:3]
```

```python
df.iloc[1:-1,1:-1]
```

```python
#select only one specific colummn
df.iloc[:,2:3]
```

```python
df.iloc[1:2,:]
```

```python
#Selecting a specific row
df.iloc[[0,2],[1,3]]
```

```python
#Pipe Function :- This method in a pandas Dataframe allows you to apply a function to the Dataframe, similar to the
#way the apply() method works. The Difference is the pipe() allows you to chain multiple operations together by 
#passing the output of one function to the input of the next function
print(df)

def add_(a,b):
    return a+b

def sub_(a,b):
    return a-b
```

```python
#Example 1 
df.pipe(add_,5)
```

```python
df.pipe(sub_,24)
```

```python
def mean_(i):
    return i.mean()

def square(i):
    return i **2 

df.pipe(mean_).pipe(square)
```

```python
#The Groupby function in pandas is used to split a dataframe into groups based on one or more columns.
#It returns a DataframeGroupBy object ,which is similar to a Dataframe but has some additional methods to perform
# operations on the grouped Data
cricket = {
    'Team':["India","India","Australia","Australia","SA","SA","SA","SA","NZ","NZ","NZ","India"],
    "Rank":[2,3,1,2,3,4,1,1,2,4,1,2],
    "Year":[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
    "Point":[876,801,891,815,776,784,834,824,758,691,883,782]
}

df=pd.DataFrame(cricket)
df

```

```python
df.groupby('Team').groups
```

```python
df.groupby(["Team","Year"]).groups
```

```python
df.groupby(["Team","Year"]).get_group(("Australia",2014))
```

```python
df.groupby("Team")["Point"].sum()
```

```python
df.groupby("Team")["Point"].sum().sort_values(ascending=False)
```

```python
groups=df.groupby(["Team","Year"])
groups.first()
```

```python
groups.get_group(('India',2015))
```

```python
groups=df.groupby("Team")
print(groups.get_group('NZ'))
print("-"*30)
print(groups.get_group('SA'))
```

```python
groups['Point'].max()
```

```python
groups = df.groupby("Team")
groups['Point'].agg([np.sum,np.mean,np.std,np.max,np.min])
```
