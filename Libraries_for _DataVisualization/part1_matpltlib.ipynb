### Generating Sin and Cos Graph using Matplotlib

```python
x=np.linspace(0,4*np.pi,100)
sin_y=np.sin(x)
cos_y=np.cos(x)
plt.plot(x,sin_y,label='Sin')
plt.plot(x,cos_y,label='Cos')
plt.legend(loc="upper right") #legend is used to add label to the graph, the loc argument takes as an integer or a
                              #a string that specifies the location of the legend on the Plot.
                              #such as lower right,center,upper right,lower left 
plt.title("Sin and Cos Waves")
```

```python
x=np.linspace(0,4*np.pi,100)
sin_y=np.sin(x)
cos_y=np.cos(x)
plt.plot(x,sin_y,'*',label='Sin')
plt.plot(x,cos_y,'--',label='Cos')
plt.legend(loc="upper right") 
plt.title("Sin and Cos Waves")
```

```python
#Adjust the Plot Axes
# can be done using plt.xlim for x-axis and plt.ylim for y-axis
x = np.linspace(0,4*np.pi,100)
plt.xlim(0,15)
plt.ylim([-2.5,2.5])
sin_y = np.sin(x)
plt.plot(x,sin_y)
plt.title('Sine wave')
#to avoid adjust graph we can use plt.axis('tight') or plt.axis('equal')
```

```python
#Using ".set" to combine all of these commands
x = np.linspace(0,4*np.pi)
fig = plt.axes()
fig.plot(x,np.sin(x))
fig.set(xlim=(0,10),ylim=(-2.5,2.5),xlabel='x',ylabel='sin(x)',title = 'sin Wave')
fig.plot()
```

### Histogram and Binnings

```python
# A histogram is a plot that displays the distribution of a set of continuous and discrete data.
data=np.random.randn(10000000)
plt.hist(data,bins=1000,edgecolor='blue',color='green',alpha=0.8)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Random Histogram")
plt.show()
```

```python
#Individual Bins can bee removed using histtype=stepfilled
data=np.random.randn(10000)
plt.hist(data,bins=100,edgecolor='blue',color='green',alpha=0.8,histtype='stepfilled')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Random Histogram")
plt.show()
```

```python
#Multiple Histogram
x1=np.random.randn(1000)
x2=np.random.randn(1000) + 3
x3=np.random.randn(1000) + 5
plt.hist(x1,bins=20,alpha=0.8)
plt.hist(x2,bins=20,alpha=0.8)
plt.hist(x3,bins=20,alpha=0.8)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Random Histogram")
plt.show()
```

```python
# 2D Histogram
# plt.hist2d is a function in matplotlib that generates a 2D Histogram.
# A 2D histogram is a representation of the distributiion of data over two variables.
# 2D histogram shows the relationship between two variables
x=np.random.randn(1000)
y=np.random.randn(1000) + 3
plt.hist2d(x,y,bins=20,cmap='viridis')
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Histogram")
plt.show()
```

```python
x=np.random.randn(1000)
y=np.random.randn(1000) + 3
plt.hist2d(x,y,bins=20,cmap='PuBuGn')
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
plt.title("2D Histogram")
plt.show()
```

```python
#Pie Chart
# A Pie chart is a circular statistical graphic, which is divided into slices to iilustrate numerical proportion.
cars=['Audi','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [450 , 540, 40, 54, 333, 369]
explode=[0, 0.2, 0, 0, .1, 0]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,wedgeprops={"edgecolor":"black"})
plt.show()
```

`Adding a shadow`

```python
cars=['Audi','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [450 , 540, 40, 54, 333, 369]
explode=[0, 0.2, 0, 0, .1, 0]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,shadow=True,wedgeprops={"edgecolor":"black"})
plt.show()
```

`adding Percentage`

```python
cars=['Audi','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [450 , 540, 40, 54, 333, 369]
explode=[0, 0.2, 0, 0, .1, 0]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%1.2f%%')
plt.show()
```

```python
# Adding Wedget
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%1.2f%%',
        wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':True})
plt.show()
```

```python
# Multiple Subplot :- refer to dividing a figure into multiple subplots or axes
x = np.linspace(0,10,50)
fig,axis = plt.subplots(2,2)
y=np.sin(x)
axis[0,0].plot(x,y)
axis[0,0].set_title("Sin waves") 

y=np.cos(x)
axis[0,1].plot(x,y)
axis[0,1].set_title("Cos waves") 

y=np.tan(x)
axis[1,0].plot(x,y)
axis[1,0].set_title("Tan waves")

y=np.random.randn(1000)
axis[1,1].hist(y,bins=10)
axis[1,1].set_title("Random DIstribution")

plt.tight_layout()
```
